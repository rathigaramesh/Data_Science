# Interquartile Range IQR (Lesser and Greater)

```python
# | Statistic | Meaning                             |
# | --------- | ----------------------------------- |
# | Mean      | Average                             |
# | Median    | 50th percentile (Q2)                |
# | Mode      | Most frequent value                 |
# | Q1:25%    | 25th percentile (lower quartile)    |
# | Q2:50%    | 50th percentile (median)            |
# | Q3:75%    | 75th percentile (upper quartile)    |
# | 99%       | 99th percentile                     |
# | Q4:100%   | Maximum                             |
# | IQR       | Interquartile Range (Q3 - Q1)       |
# | 1.5rule   | 1.5 Ã— IQR (used for outlier bounds) |
# | Lesser    | Q1 - 1.5Ã—IQR (lower outlier limit)  |
# | Greater   | Q3 + 1.5Ã—IQR (upper outlier limit)  |
# | Min       | Minimum value                       |
# | Max       | Maximum value                       |
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]
    descriptive[columnName]["Lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Min"]=dataset[columnName].min()
    descriptive[columnName]["Max"]=dataset[columnName].max()
    
```

```python
def freqTable(columnName,dataset):
    freqTable=pd.DataFrame(columns=["Unique_Values","Frequency","Relative Frequency","Cusum"])
    freqTable["Unique_Values"]=dataset[columnName].value_counts().index
    freqTable["Frequency"]=dataset[columnName].value_counts().values
    freqTable["Relative Frequency"]=(freqTable["Frequency"]/103)
    freqTable["Cusum"]=freqTable["Relative Frequency"].cumsum()
    return freqTable
```

```python
def Univariate_freq(dataset,quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
        descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
        descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["Lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
        descriptive[columnName]["Greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["Max"]=dataset[columnName].max()
    return descriptive
```

```python
def Univariate_histSkewKuto(dataset,quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater",
                                    "Min","Max","kurtosis","skew"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
        descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
        descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["Lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
        descriptive[columnName]["Greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["Max"]=dataset[columnName].max()
        descriptive[columnName]["kurtosis"]=dataset[columnName].kurtosis()
        descriptive[columnName]["skew"]=dataset[columnName].skew()
    return descriptive
```

```python
from scipy.stats import skew, kurtosis

# Create an empty DataFrame with statistical rows
descriptive = pd.DataFrame(index=[
    "Count", "Mean", "Median", "Mode",
    "Min", "Max", "Range",
    "Q1:25%", "Q2:50%", "Q3:75%", "99%", "Q4:100%",
    "IQR", "1.5rule", "Lesser", "Greater",
    "Std", "Variance", "Skewness", "Kurtosis", "Sum"
])

# Loop over numerical columns
for columnName in quan:
    col = dataset[columnName].dropna()  # drop NaNs to avoid errors

    descriptive.at["Count", columnName] = col.count()
    descriptive.at["Mean", columnName] = col.mean()
    descriptive.at["Median", columnName] = col.median()
    descriptive.at["Mode", columnName] = col.mode().iloc[0] if not col.mode().empty else np.nan
    descriptive.at["Min", columnName] = col.min()
    descriptive.at["Max", columnName] = col.max()
    descriptive.at["Range", columnName] = col.max() - col.min()
    descriptive.at["Q1:25%", columnName] = col.quantile(0.25)
    descriptive.at["Q2:50%", columnName] = col.quantile(0.50)
    descriptive.at["Q3:75%", columnName] = col.quantile(0.75)
    descriptive.at["99%", columnName] = np.percentile(col, 99)
    descriptive.at["Q4:100%", columnName] = col.max()
    descriptive.at["IQR", columnName] = col.quantile(0.75) - col.quantile(0.25)
    descriptive.at["1.5rule", columnName] = 1.5 * descriptive.at["IQR", columnName]
    descriptive.at["Lesser", columnName] = descriptive.at["Q1:25%", columnName] - descriptive.at["1.5rule", columnName]
    descriptive.at["Greater", columnName] = descriptive.at["Q3:75%", columnName] + descriptive.at["1.5rule", columnName]
    descriptive.at["Std", columnName] = col.std()
    descriptive.at["Variance", columnName] = col.var()
    descriptive.at["Skewness", columnName] = skew(col)
    descriptive.at["Kurtosis", columnName] = kurtosis(col)
    descriptive.at["Sum", columnName] = col.sum()
```
